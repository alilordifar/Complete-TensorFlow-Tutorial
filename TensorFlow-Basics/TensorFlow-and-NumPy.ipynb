```python
import tensorflow as tf
import numpy as np
```

### Tensors and NumPy
TensorFlow interacts beautifully with NumPy arrays.

```python
# Create a tensor directly from NumPy array
A = tf.constant(np.array([1, 2, 3]))
A
```

```python
# Convert a tensor to NumPy array
np.array(A), type(np.array(A))
```

```python
# Convert tensor A to a NumPy array
A.numpy(), type(A.numpy())
```

```python
# The default type of each are slightly different 
numpy_A = tf.constant(np.array([4., 2., 5.]))
tensor_A = tf.constant([4., 2., 5.])
# Cehck the datatype of each
numpy_A.dtype, tensor_A.dtype
```
