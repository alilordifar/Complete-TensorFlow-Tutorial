### Aggregating Tensors

Considering tensors from multiple values down to a smaller amount of values.

```python
# Getting the absolute values
A = tf.constant([-3, -8])
A
```

```python
# Get the absolute values
tf.abs(A)
```

Let's go through the following forms of aggregation:
* Get the minimum
* get the maximum
* get the mean of a tensor
* get the sum of a tensor

```python
# Create a random tensor with values between 0 and 10 of size 50
B = tf.random.Generator.from_seed(42)
B = B.normal((3, 2))
B
```

```python
tf.math.reduce_min(B)
```

```python
tf.math.reduce_max(B)
```

```python
tf.math.reduce_mean(B)
```

```python
import numpy as np
E = tf.constant(np.random.randint(0, 100, size = 50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the max
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# To find the variance of our tensor we need access to tensorflow probability
#Using tfp --> tensorflow probability
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# Find the standard deviation (std)
tf.math.reduce_std(tf.cast(E, dtype = tf.float32))
```

```python
tf.math.reduce_std(E)
```

#### Find the positional maximum and minimum
* At which index of a tensor or of a row, does the maxiumum value occur. It is usefull when you trainthe model and want to find the maximum probability in a certain index.
* In other words, if we want to find the maximum element position of G meaning at which index does the maximum value occur. 


```python
# Create a new tensor for finding positionalminimum and maximum
tf.random.set_seed(42)
G = tf.random.uniform(shape = [50])
G
```

```python
# Find the positional maximum
tf.argmax(G)
```

```python
# Index on our largest value position
G[tf.argmax(G)]
```

```python
tf.reduce_max(G)
```

```python
# Check for equality 
assert G[tf.argmax(G)] == tf.reduce_max(G) # With assert we will get error if it's False
```

```python
G[tf.argmax(G)] == tf.reduce_max(G)
```

```python
# Find the positional minimum
tf.argmin(G)
```

```python
# Find the minimum using the positional minimum index
G[tf.argmin(G)]
```
