## Squaring, log, square root

```python
import tensorflow as tf
```

```python
# Create a new tensor
H = tf.range(1, 10)
H
```

```python
# Square tensor
tf.square(H)
```

```python
# Square root
tf.sqrt(tf.cast(H, dtype = tf.float32)) # Try it without specify the dtype
```

```python
# Find the log
tf.math.log(tf.cast(H, dtype = tf.float32))
```
